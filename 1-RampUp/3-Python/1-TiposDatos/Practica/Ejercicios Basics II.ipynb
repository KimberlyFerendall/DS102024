## Ejercicio 5
Crea un programa donde se recojan dos inputs del usuario, y el output del programa sea si esos inputs son iguales o no

````python
libros_espanhol = int(input("¿Cuántos libros has leído en español en el último mes?"))
libros_ingles = int(input("¿Cuántos libros has leído en inglés?"))
if (libros_espanhol == libros_ingles):
    print("Has leído la misma cantidad de libros en ambos idiomas")
elif (libros_espanhol != libros_ingles):
    print("Has leído diferentes cantidades de libros en ambos idiomas")
````

## Ejercicio 7
Crea un programa que recoja dos inputs. Tiene que comprobar si su suma es igual, superior o inferior a 10

## Ejercicio 10
Para el siguiente string se pide imprimir por pantalla los siguientes casos

    "A quien madruga, dios le ayuda"

1. Pasarlo todo a mayusculas
2. Pasarlo todo a minusculas
3. Pasar a mayusculas solo las iniciales de las palabras
4. Crea una lista dividiendolo por sus espacios
5. Sustituye las comas `,` por puntos y comas `;`
6. Elimina las `a` minusculas

````python
string_ejemplo = "A quien madruga, dios le ayuda"
print(string_ejemplo.upper())
print(string_ejemplo.lower())
print(string_ejemplo.title())
print(string_ejemplo.split())
print(string_ejemplo.replace("," , ";"))
print(string_ejemplo.replace("a", ""))
````
